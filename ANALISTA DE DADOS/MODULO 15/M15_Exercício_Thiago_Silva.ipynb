## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[["cut", "price", "carat"]]
diamonds.head()

with sns.axes_style('whitegrid'):

  grafico = sns.scatterplot(data=diamonds, x="carat", y="price", hue="cut", palette="pastel")
  grafico.set(title='Relação Preço e peso por Corte:', xlabel='Peso', ylabel='Preço');
  grafico.get_legend().set_title("Tipo de Corte");
```

**Insight do gráfico 1**:

Temos nesse gráfico que o nosso outliers estão localizados no tipo de corte 'Fair" a partir do valores de aproximados 6000.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2
diamonds = data[["price", "carat", "clarity"]]
diamonds.head()

with sns.axes_style('whitegrid'):

  grafico = sns.scatterplot(data=diamonds, x="price", y="carat", hue="clarity", palette="pastel")
  grafico.set(title='Relação Peso e Preço por Transparência:', xlabel='Preço', ylabel='Peso');
  grafico.get_legend().set_title("Transparência");
```

**Insight do gráfico 2**:

Apartir desse grafico, podemos identificar que a Transparência 'IF' possui o valor mais caro, levando em conta o fato de ter o menor preço.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
#print(data['name'].drop_duplicates())
data = data.loc[data['name']=='honda civic']
```

```
# gráfico do exercício 3
#Afim de ter uma melhor visualização, vou selecionar carroes especificos carros específicos:


with sns.axes_style('whitegrid'):

  cars =  sns.boxplot(x=data["weight"], y=data["name"])
  cars.set(title="Alteração do peso do Honda civic:", xlabel="Peso", ylabel="Nome");
```

**Insight do gráfico 3**:

Podemos verificar a alteração do peso do carro, tentendo realizando um análise,afirmo que o carro ficou mais leve com o passar do tempo. Digo isso levando em conta, uma tentativa do mercado de deixar os carros mais econômicos.
